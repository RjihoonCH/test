```python
import requests
import pandas as pd
import xmltodict
import json

years = ['2016','2022']
date = []
temp = []
max_temp = []
min_temp = []
hum = []
widdir = []
wind = []
rain = []
sun_Qy = []
idx= 12

for year in years:
    url = 'http://apis.data.go.kr/1390802/AgriWeather/WeatherObsrInfo/GnrlWeather/getWeatherYearMonList?serviceKey=zpwPdE1deGAf1VAFZKWybmt2YlCt%2FmMtzBnjlCRnJegpjiqpDrcjTItEZOKNLrQnv0x%2BMkFxvp6M7lB%2BnFvQyg%3D%3D&Page_No=1&Page_Size=20&search_Year='+year+'&obsr_Spot_Code=699946A001'
    response = requests.get(url)
    preparsing = xmltodict.parse(response.text)
    parsing = json.dumps(preparsing) 
    json_object= json.loads(parsing)
    dataOfMonth = json_object['response']['body']['items']['item']

    for i in range(6,idx):
        date.append(dataOfMonth[i]['date'])

    for i in range(6,idx):
        temp.append(dataOfMonth[i]['temp'])

    for i in range(6,idx):
        max_temp.append(dataOfMonth[i]['max_Temp'])

    for i in range(6,idx):
        min_temp.append(dataOfMonth[i]['min_Temp'])
    
    for i in range(6,idx):
        hum.append(dataOfMonth[i]['hum'])

    for i in range(6,idx):
        rain.append(dataOfMonth[i]['rain'])
    
    for i in range(6,idx):
        sun_Qy.append(dataOfMonth[i]['sun_Qy'])
        

elements = {'측정월':date, '기온(°C)':temp, '최고 기온(°C)':max_temp, '최저 기온(°C)':min_temp, '습도(%)':hum,
            '강수량(mm)':rain, '일사량(MJ/m²)':sun_Qy}

for key in elements.keys():
    for x in range(0,len(date)):
        if elements[key][x]==None or elements[key][x]=='0':
            elements[key][x] = elements[key][x-6]
elements
```
